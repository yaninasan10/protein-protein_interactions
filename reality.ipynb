```python
import pandas as pd
import numpy as np

new_db = pd.read_csv('~/protein-protein_interactions/FINAL_SEC_STR')
```

```python
new_db
```

```python
d = []
for i in range(0, 1814): 
    if [new_db['UNIPROT_SYMBOL_A'][i], new_db['UNIPROT_SYMBOL_B'][i]] not in d:
        d.append([new_db['UNIPROT_SYMBOL_A'][i], new_db['UNIPROT_SYMBOL_B'][i]]) 
```

```python
len(d)
```

```python
lst = []
for i in range(0, 1258):
    for j in range(0, 1814):
        if new_db['UNIPROT_SYMBOL_A'][j] == d[i][0] and new_db['UNIPROT_SYMBOL_B'][j] == d[i][1]:
            lst.append(j)
            break
    
```

```python
len(lst)
```

```python
new = new_db.loc[lst]
```

```python
new = new.reset_index(drop=True)
new
```

```python
X_1 = new.drop(['class', 'UNIPROT_SYMBOL_A', 'UNIPROT_SYMBOL_B'], axis=1)
y_1 = new['class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size = 0.2, random_state = 42)
def rf(X_train, y_train):
    clf = RandomForestClassifier(random_state=42)
    par = {'n_estimators' : range(5, 45), 'max_depth' : range(5, 25)}
    gr = GridSearchCV(clf, par, cv=5) 
    gr.fit(X_train, y_train)
    return gr
```

```python
gr_1 = rf(X_train_1, y_train_1)
```

```python
gr_1.best_params_
```

```python
best_clf_1 = gr_1.best_estimator_
```

```python
y_pred_1 = best_clf_1.predict(X_test_1)
```

```python
best_clf_1.score(X_test_1, y_test_1)
```
