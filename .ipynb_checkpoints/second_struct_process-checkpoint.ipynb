После работы с Jpred API имеем вторичные структуры белков из рассматриваемой таблицы. Закодируем их, сопоставив C(-), E, H соответственно 1, 2 и 3. Также сделаем их всех одинаковой длины, приняв за длину по умолчанию наибольшую длину и дополнив до нее нулями.

```python
dir_1 = '/home/yanina/protein-protein_interactions/sec_str_a/'
dir_2 = '/home/yanina/protein-protein_interactions/sec_str_a_dig'
max_size_1 = 796
dir_3 = '/home/yanina/protein-protein_interactions/sec_str_b/'
dir_4 = '/home/yanina/protein-protein_interactions/sec_str_b_dig'
max_size_2 = 798

def transformIntoDigits(dir_, dir_2, max_size):
    for filename in os.listdir(dir_):
        text = open(dir_ + filename).read()
        new_text = text.replace('-', '1')
        new_text = new_text.replace('E', '2')
        new_text = new_text.replace('H', '3')
        new_text = new_text[:-2]
        while len(new_text) < max_size:
            new_text = new_text + '0'
        with open(os.path.join(dir_2, filename), 'w') as file:
            file.write(new_text)
            
            
transformIntoDigits(dir_1, dir_2, max_size_1)
transformIntoDigits(dir_3, dir_4, max_size_2)
```
