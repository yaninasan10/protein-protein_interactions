Теперь закодируем последовательности аминокислот, используя conjoint triad method.

```python
def frequency(seq):
    sequence = [0 for l in range (0, 343)]
    for i in range(0, (len(seq) - 3)):
        subSeq = seq[i:i+3]
        tmp = ''
        for j in range(0,3):
            if((subSeq[j] == 'A') or (subSeq[j] == 'G') or (subSeq[j] == 'V')):
                tmp += "1"
            elif((subSeq[j] == 'Y') or (subSeq[j] == 'M') or (subSeq[j] == 'T') or (subSeq[j] == 'S')):
                tmp += "2"
            elif((subSeq[j] == 'I') or (subSeq[j] == 'L') or (subSeq[j] == 'F') or (subSeq[j] == 'P')):
                tmp += "3"
            elif((subSeq[j] == 'H') or (subSeq[j] == 'N') or (subSeq[j] == 'Q') or (subSeq[j] == 'W')):
                tmp += "4"
            elif((subSeq[j] == 'R') or (subSeq[j] == 'K')):
                tmp += "5"
            elif((subSeq[j] == 'D') or (subSeq[j] == 'E')):
                tmp += "6"
            elif((subSeq[j] == 'C')):
                tmp += "7"
            else:
                return []                
        ind = 49 * (int(tmp[0]) - 1) + 7 * (int(tmp[1]) - 1) + int(tmp[2]) - 1
        sequence[ind] += 1
    return sequence
```

```python
list_seq_a = np.array([frequency(x) for x in database['SEQUENCE_A']])
list_seq_b = np.array([frequency(x) for x in database['SEQUENCE_B']])
```

```python
database['CODE_SEQ_A'] = [x for x in list_seq_a]
database['CODE_SEQ_B'] = [x for x in list_seq_b]
```

Разобьем теперь закодированную последовательность на 343 отдельные фичи. Сложности возникают из-за того, что после сохранения таблицы,чтобы ничего не потерять, и повторного чтения, массивы последовательностей превращаются в строки и их надо предварительно парсить. Также не хватает оперативной памяти, что отделить все колонки до конца, поэтому это дело делалось при помощи Google Colab, а здесь можно увидеть используемый код.

```python
database_new['CODE_SEQ_A'] = database_new['CODE_SEQ_A'].apply(lambda x:  np.fromiter(map(int, x.strip('[]').split(', ')), dtype = np.int))
database_new['CODE_SEQ_B'] = database_new['CODE_SEQ_B'].apply(lambda x:  np.fromiter(map(int, x.strip('[]').split(' ')), dtype = np.int))
```

```python
for i in range(0, 343):
    print(i)
    database_new["seq_a_" + str(i)] = [x[i] for x in database_new['CODE_SEQ_A']]
```

```python
for i in range(0, 343):
    print(i)
    database_new["seq_b_" + str(i)] = [x[i] for x in database_new['CODE_SEQ_B']]
```
